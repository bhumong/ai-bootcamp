$$

-2 & 5 & -5 & 3 & 1 \\

M \cdot X =
 =

 -40

What is the value of $a + b + c + d + e$?

```python
!pip install numpy
```

```python
# You may add any code here to derive your variables
# Please change this
import numpy as np
M = np.array([
    [-2, 3, 3, 3, -4],
    [0, -4, -1, 1, 2],
    [1, 5, 4, 2, 0],
    [-2, 5, -5, 3, 1],
    [-3, 2, 4, 3, 4]
])
b = np.array([-33, 9, -34, -38, -40])
X = np.linalg.solve(M, b)
X = np.rint(X).astype(int)
print(X)

a = X[0]
b = X[1]
c = X[2]
d = X[3]
e = X[4]
total = a + b + c + d + e

print(f"The total is {total}")

assignment_id = "09-deep-learning-project"
question_id = "q1_linear_algebra"
submit(student_id, name, assignment_id, str(total), question_id, drive_link)
```

## Task 3

Given the following Neural Networks:

First layer:

$$
W = \begin{bmatrix}
0.23 & 0.67 & 0.12 \\
-0.89 & -0.45 & 0.78 \\
0.34 & 0.56 & -0.90 \\
-0.12 & 0.34 & 0.56 \\
0.78 & -0.90 & 0.23 \\
\end{bmatrix}
$$

$$
bias = \begin{bmatrix}
0.23 \\
-0.89 \\
0.34 \\
0.12 \\
-0.78 \\
\end{bmatrix}
$$

Second layer:

$$
W = \begin{bmatrix}
0.23 & 0.67 & 0.12 & 0.45 & 0.89 \\
0.12 & 0.34 & 0.56 & 0.78 & 0.90 \\
\end{bmatrix}
$$

$$
bias = \begin{bmatrix}
1.96 \\
-1.08 \\
\end{bmatrix}
$$

Third layer:
$$
W = \begin{bmatrix}
1.08 & -0.16
\end{bmatrix}
$$

$$
bias = \begin{bmatrix}
-2.8
\end{bmatrix}
$$

All layers use the ReLU activation function.

What is the output given the following inputs?

$$
X = \begin{bmatrix}
1 \\
2 \\
4 \\
\end{bmatrix}



```python
# You may add any code here to derive your variables
import numpy as np

def relu(x):
    return np.maximum(0, x)

# input
X = np.array([1, 2, 4])

# 1st layer
W1 = np.array([
    [0.23, 0.67, 0.12],
    [-0.89, -0.45, 0.78],
    [0.34, 0.56, -0.90],
    [-0.12, 0.34, 0.56],
    [0.78, -0.90, 0.23]
])
b1 = np.array([0.23, -0.89, 0.34, 0.12, -0.78])

# 2nd layer
W2 = np.array([
    [0.23, 0.67, 0.12, 0.45, 0.89],
    [0.12, 0.34, 0.56, 0.78, 0.90]
])
b2 = np.array([1.96, -1.08])

# 3rd layer
W3 = np.array([1.08, -0.16])
b3 = -2.8

# result 1st layer
Z1 = relu(np.dot(W1, X) + b1)

# result 2nd layer
Z2 = relu(np.dot(W2, Z1) + b2)

# Please change this
# output
output = round(relu(np.dot(W3, Z2) + b3), 2)

print(f"The output is {output}")

assignment_id = "09-deep-learning-project"
question_id = "q3_complex_neural_network"
submit(student_id, name, assignment_id, str(output), question_id, drive_link)
```
