```python
import pandas

# Drop unused column
drop_column = ['ID', 'Photo', 'Flag', 'Club Logo', 'Loaned From', 'Name', 'Jersey Number', 'Joined', 'Contract Valid Until', 'Nationality', 'Club', 'Body Type', 'Real Face', 'Marking']
df = pandas.read_csv("./FIFA21_official_data.csv")
df.drop(columns=drop_column, inplace=True)

df
```

```python
# parse money
def parse_money(val):
    val_str = f"{val}".replace("€", "")
    if "K" in val_str:
        val_str = val_str.replace("K", "")
        return float(val_str) * 1000
    elif "M" in val_str:
        val_str = val_str.replace("M", "")
        return float(val_str) * 1_000_000
    return 0
df["Value"] = df["Value"].map(parse_money)
df["Wage"] = df["Wage"].map(parse_money)
df["Release Clause"] = df["Release Clause"].map(parse_money)
```

```python
# Encode Preferred Foot
df['Preferred Foot'] = df['Preferred Foot'].map({'Left': 0, 'Right': 1})
df['Preferred Foot']
```

```python
# Encode Work Rate
df[['Work Rate Attack', 'Work Rate Defense']] = df['Work Rate'].str.split('/ ', expand=True)
df["Work Rate Attack"] = df["Work Rate Attack"].str.replace(" ", "")
df["Work Rate Defense"] = df["Work Rate Defense"].str.replace(" ", "")
work_rate_mapping = {'High': 2, 'Medium': 1, 'Low': 0}
df['Work Rate Attack'] = df['Work Rate Attack'].map(work_rate_mapping)
df['Work Rate Defense'] = df['Work Rate Defense'].map(work_rate_mapping)
df = df.drop(columns=['Work Rate'])
df[["Work Rate Attack", "Work Rate Defense"]]
```

```python
# Encode Position and Best Position
df['Position'] = df['Position'].str.replace(r'<.*?>', '', regex=True)
df['Best Position'] = df['Best Position'].str.replace(r'<.*?>', '', regex=True)
df = pandas.get_dummies(df, columns=['Position', 'Best Position'], prefix=['Pos', 'BestPos'])
```

```python
# Convert Height
df['Height'] = (
    df['Height']
    .str.split("'", expand=True)
    .apply(lambda col: col.astype(int))
    .apply(lambda x: round((x[0] * 30.48) + (x[1] * 2.54)), axis=1)
)

# Convert Weight
df['Weight'] = (
    df['Weight']
    .str.replace('lbs', '')
    .astype(int)
    .apply(lambda x: round(x / 2.2046))
)
```

```python
df.head(5).to_csv("./export.csv")
```

```python
df['OK Player'] = (df['Overall'] >= 70)
```

```python
print("Total NaNs in original data:", df.isnull().sum().sum())
print("Columns with NaNs:")
print(df.isnull().sum()[df.isnull().sum() > 0])
df.shape[0]
```

```python
df_copy = df.copy()
df.drop(columns=['Overall', 'Best Overall Rating'], inplace=True)
df.columns
```

```python
from sklearn.model_selection import train_test_split

y = df['OK Player']
x = df.drop(columns=["OK Player"])

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```

```python
from sklearn.impute import SimpleImputer

# Impute Na missing data (null)
num_imputer = SimpleImputer(strategy='median')
X_train_imputed = num_imputer.fit_transform(X_train)
X_test_imputed = num_imputer.transform(X_test)
```

```python
# Use Standar Scalar before use SVM model
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.fit_transform(X_test_imputed)
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', class_weight='balanced', probability=True)
svm_model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = svm_model.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
