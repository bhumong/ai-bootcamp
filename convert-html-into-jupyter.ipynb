```python
import os
from urllib.parse import urlparse
def create_folder_from_url(url, base_dir="learn"):
    # Parse the URL
    parsed_url = urlparse(url)
    path = parsed_url.path  # Get the path part of the URL

    # Find the section after '/learn/' and before the last segment
    parts = path.split("/learn/")[1]  # Get the part after '/learn/'
    folder_path = os.path.join(base_dir, os.path.dirname(parts))  # Remove the .html file

    # Create the directory
    os.makedirs(folder_path, exist_ok=True)
    print(f"Created folder: {folder_path}")
    return f"{folder_path}"
```

```python
import re
import base64

images_dir = "notebook_images"  # Directory to save images

# Create a directory for images
if not os.path.exists(images_dir):
    os.makedirs(images_dir)

for url in links:
    print(url)
    response = requests.get(url)
    response.raise_for_status()  # Ensure the request was successful

    soup = BeautifulSoup(response.text, 'html.parser')

    # Adjust the selector to target the main content area.
    content_div = soup.find('main', class_='content')
    if content_div is None:
        raise Exception("Content container not found.")

    folder = create_folder_from_url(url)
    filename = ""
    try:
        filename = url.split('/')[-1].split('.')[0]
    except Exception:
        pass

    notebook = nbf.v4.new_notebook()

    if content_div:
        notebook.cells.append(nbf.v4.new_markdown_cell(f"source: [link]({url})"))
        for element in content_div.find_all(["h1", "h2", "h3", "h4", "p", "pre", "img", "table", "ul"]):
            if element.name in ["h1", "h2", "h3", "h4"]:
                # Convert headers to markdown cells
                notebook.cells.append(nbf.v4.new_markdown_cell(f"# {element.get_text()}"))
            elif element.name == "p":
                # Convert paragraphs to markdown cells
                notebook.cells.append(nbf.v4.new_markdown_cell(element.get_text()))
            elif element.name == "pre":
                # Convert code blocks to code cells
                code_text = element.get_text()
                notebook.cells.append(nbf.v4.new_code_cell(code_text))
            elif element.name == "table":
                # Convert tables to markdown
                table_md = []
                rows = element.find_all("tr")

                for row_index, row in enumerate(rows):
                    columns = row.find_all(["th", "td"])
                    row_text = " | ".join(col.get_text(strip=True) for col in columns)
                    table_md.append(f"| {row_text} |")

                    # Add separator after header row
                    if row_index == 0:
                        table_md.append("|" + " --- |" * len(columns))

                # Convert list to markdown table string
                table_md_str = "\n".join(table_md)
                notebook.cells.append(nbf.v4.new_markdown_cell(table_md_str))
            elif element.name == "ul":
                # Convert unordered lists (<ul><li>) into markdown bullet lists
                list_items = ["- " + li.get_text(strip=True) for li in element.find_all("li")]
                list_md = "\n".join(list_items)
                notebook.cells.append(nbf.v4.new_markdown_cell(list_md))
            elif element.name == "img":
                # Process images (including Base64 encoded ones)
                img_src = element.get("src")
                
                if img_src and img_src.startswith("data:image"):
                    # Extract Base64 data
                    match = re.match(r"data:image/(?P<ext>[a-z]+);base64,(?P<data>.+)", img_src)
                    if match:
                        img_ext = match.group("ext")
                        img_data = match.group("data")
                        img_folder = f"{folder}/{images_dir}"
                        if not os.path.exists(img_folder):
                            os.makedirs(img_folder)

                        # Decode and save image
                        img_filename = f"image_{len(os.listdir(images_dir))}.{img_ext}"
                        img_path = os.path.join(img_folder, img_filename)

                        with open(img_path, "wb") as img_file:
                            img_file.write(base64.b64decode(img_data))

                        # Add Markdown cell with image reference
                        markdown_img = f"![Image](./{images_dir}/{img_filename})"
                        notebook.cells.append(nbf.v4.new_markdown_cell(markdown_img))

                if img_src and img_src.startswith("http"):
                    markdown_img = f"![Image]({img_src})"
                    notebook.cells.append(nbf.v4.new_markdown_cell(markdown_img))



    # Save the Jupyter notebook
    notebook_path = f"{folder}/{filename}-generated.ipynb"
    with open(notebook_path, "w", encoding="utf-8") as f:
        nbf.write(notebook, f)

    
    
    
```

import os
import requests
from bs4 import BeautifulSoup
import nbformat as nbf
from urllib.parse import urljoin

# Define input and output paths
html_path = "your_quarto_file.html"  # Change this to your actual HTML file
notebook_path = "converted_notebook.ipynb"
images_dir = "notebook_images"  # Directory to save images

# Create a directory for images
if not os.path.exists(images_dir):
    os.makedirs(images_dir)

# Load the HTML file
with open(html_path, "r", encoding="utf-8") as file:
    soup = BeautifulSoup(file, "html.parser")

# Create a new Jupyter notebook
notebook = nbf.v4.new_notebook()

# Extract the main content
content_blocks = soup.find("main", {"id": "quarto-document-content"})

if content_blocks:
    for element in content_blocks.find_all(["h1", "h2", "h3", "h4", "p", "pre", "img"]):
        if element.name in ["h1", "h2", "h3", "h4"]:
            # Convert headers to markdown cells
            notebook.cells.append(nbf.v4.new_markdown_cell(f"# {element.get_text()}"))
        elif element.name == "p":
            # Convert paragraphs to markdown cells
            notebook.cells.append(nbf.v4.new_markdown_cell(element.get_text()))
        elif element.name == "pre":
            # Convert code blocks to code cells
            code_text = element.get_text()
            notebook.cells.append(nbf.v4.new_code_cell(code_text))
        elif element.name == "img":
            # Process images
            img_src = element.get("src")
            if img_src:
                # Extract filename and save the image
                img_filename = os.path.basename(img_src)
                img_path = os.path.join(images_dir, img_filename)

                # Download the image (if it's an online image)
                if img_src.startswith("http"):
                    response = requests.get(img_src)
                    with open(img_path, "wb") as img_file:
                        img_file.write(response.content)
                else:
                    # If it's a local reference, copy it to the output directory
                    img_src = os.path.join(os.path.dirname(html_path), img_src)
                    if os.path.exists(img_src):
                        with open(img_src, "rb") as img_file, open(img_path, "wb") as out_file:
                            out_file.write(img_file.read())

                # Add Markdown cell with an image reference
                markdown_img = f"![Image](./{img_path})"
                notebook.cells.append(nbf.v4.new_markdown_cell(markdown_img))

# Save the Jupyter notebook
with open(notebook_path, "w", encoding="utf-8") as f:
    nbf.write(notebook, f)

print(f"Conversion complete! Jupyter Notebook saved as {notebook_path}")
